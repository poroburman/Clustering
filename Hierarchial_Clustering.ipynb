```python
# The purpose of this script is to find a pattern amongst the customers visiting a Mall.
# The dataset used in this script consists of customers data collected by a Mall
# Each row consists of a customer. Each column consists of a customer's variable, including his/her spending score.
```

```python
# 1. Data pre-processing
```

```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 15,10

# import the dataset
dataset = pd.read_csv('Mall_Customers.csv')

# set the dependent variables
X = dataset.iloc[:,[3,4]].values
```

```python
# print top 5 dataset values
dataset.head()
```

```python
# 2. select number of clusters
```

```python
# import hierarchy as sch
import scipy.cluster.hierarchy as sch

# create dendogram instance
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))

# plot the dendogram
plt.title("Dendogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean distances")
plt.show()
```

```python
# 3. Training Hierachial clustering model on the dataset
```

```python
# import AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering

# create an instance for AgglomerativeClustering
hierarchial_clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')

# train the instance on the data
y_hierarchial_clustering = hierarchial_clustering.fit_predict(X)

# visualize the different clusters created
print(y_hierarchial_clustering)
```

```python
# 4. Visualize the clusters
```

```python
# scatter plot for cluster 1
plt.scatter(X[y_hierarchial_clustering == 0,0], X[y_hierarchial_clustering == 0,1], s = 100, c = 'red', label = 'Cluster 1')

# scatter plot for cluster 2
plt.scatter(X[y_hierarchial_clustering == 1,0], X[y_hierarchial_clustering == 1,1], s = 100, c = 'blue', label = 'Cluster 2')

# scatter plot for cluster 3
plt.scatter(X[y_hierarchial_clustering == 2,0], X[y_hierarchial_clustering == 2,1], s = 100, c = 'green', label = 'Cluster 3')

# scatter plot for cluster 4
plt.scatter(X[y_hierarchial_clustering == 3,0], X[y_hierarchial_clustering == 3,1], s = 100, c = 'cyan', label = 'Cluster 4')

# scatter plot for cluster 5
plt.scatter(X[y_hierarchial_clustering == 4,0], X[y_hierarchial_clustering == 4,1], s = 100, c = 'magenta', label = 'Cluster 5')

# title for the plot
plt.title('Clusters of customers')

# Axes labels
plt.xlabel("Annual Income (k$)")
plt.ylabel('Spending Score (1-100)')

# show the legend
plt.legend()

# show the plot
plt.show()
```

```python
# As you can see, different clusters show the customers who have different annual incomes and their related spending scores.
# This can be used to target certain specific customers based upon these clusters.
```
